```python
#import libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
```

There are multiple strategies to handle missing data

- Removing all rows or all columns containing missing data.
- Filling all missing values with a value (mean in continouos or mode in categorical for example).
- Filling all missing values with an algorithm.

For this exercise, because we have such low amount of null values, we will drop rows containing any missing value. 

```python
# Se van a eliminar todas las columnas con valors nuelos.
spaceship.dropna(axis = 1, inplace = True)
```

**KNN**

K Nearest Neighbors is a distance based algorithm, and requeries all **input data to be numerical.**

Let's only select numerical columns as our features.

```python
# Columnas numericas: Age, RoomService, FoodCourt, ShoppingMall, Spa, VRDeck
# Se va a crear un df con ellas en concreto:
num_spaceship = spaceship[['Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']]
```

And also lets define our target.

```python
# He definido como objetivo la columna de edad 
```

**Train Test Split**

Now that we have split the data into **features** and **target** variables and imported the **train_test_split** function, split X and y into X_train, X_test, y_train, and y_test. 80% of the data should be in the training set and 20% in the test set.

```python
caracteristicas = num_spaceship.drop(columns = ["Age"])
objetivo = num_spaceship['Age']
```

```python
# Esta línea divide el df en dos conjuntos de datos:
# Uno con las caracteristicas del entrenamiento: X_train
# Otro de entrenamiento: y_train
# "caracteristicas" Es el conjunto de datos que contiene las características o variables independiente, 
# mientras que "Objetivo" es lo que quieres predecir, ya sea un problema de regresion (var´s num) o de clasificacion (var´s cat)
# test_size=0.20: Define el tamaño del conjunto de prueba. random_state=0: Este parámetro asegura que la división 
# de los datos sea reproducible. #
X_train, X_test, y_train, y_test = train_test_split(caracteristicas, objetivo, test_size=0.20, random_state=0)
```

**Model Selection**

In this exercise we will be using **KNN** as our predictive model.

You need to choose between **Classificator** or **Regressor**. Take into consideration target variable to decide.

Initialize a KNN instance without setting any hyperparameter.

```python
# Se inicia Knn y se buscan 6 vecinos. Al estar usando una variable numerica estamos trabajando en el tipo regresion, 
# de lo contrario seria tipo clasificacion
knn = KNeighborsRegressor(n_neighbors = 6)
```

Fit the model to your data.

```python
# Se inicia el entrenamiento. fit ajustara el modelo con datos de entrada (X_tain, valores independientes) 
# y de salida (Y_train, valores dependientes, vamos, los datos a predecir )
knn.fit(X_train, y_train)
```

Evaluate your model.

```python
# Vamos a hacer una pequeña evaluacion
knn.score(X_test, y_test)
```
